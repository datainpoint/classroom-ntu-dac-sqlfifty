## 注意所選擇的資料庫與儲存在其中的實體資料表

![](sqlite_studio.png)

````sqlite3
SELECT *
  FROM top_rated_movies;
````

## 資料庫與儲存在其中的實體資料表：`nba` 資料庫

- `career_summaries` 球員的生涯攻守數據，數據截至 2021-01-31。
- `players` 球員的基本資料，數據截至 2021-01-31。
- `teams` 球隊的基本資料，數據截至 2021-01-31。

## 資料庫與儲存在其中的實體資料表：`covid19` 資料庫

- `daily_report` 新冠肺炎的每日報告，數據截至 2021-01-31。
- `lookup_table` 國家的基本資訊，數據截至 2021-01-31。
- `time_series` 新冠肺炎的時間序列資料，數據截至 2021-01-31。

## 資料庫與儲存在其中的實體資料表：`imdb` 資料庫

- `actors` 演員的姓名，數據截至 2021-01-31。
- `casting` 電影與演員的資訊，數據截至 2021-01-31。
- `top_rated_movies` IMDb 最高評價的 250 部電影，數據截至 2021-01-31。

## 資料庫與儲存在其中的實體資料表：`twElection2020` 資料庫

- `legislative_at_large` 台灣 2020 不分區立委的各投票所得票明細。
- `legislative_regional` 台灣 2020 分區立委的各投票所得票明細。
- `presidential` 台灣 2020 總統副總統的各投票所得票明細。
